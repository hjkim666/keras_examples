```python
#Y one-hot encoding 

from keras.utils import to_categorical

one_hot_train_labels = to_categorical(train_labels)
one_hot_test_labels = to_categorical(test_labels)
```

```python
train_labels
```

```python
one_hot_train_labels.shape
```

```python
# x  
import numpy as np 

def vectorize_seq(seq, dimension=10000):
  results = np.zeros((len(seq), dimension))
  for i, s in enumerate(seq):
    results[i, s] = 1.
  return results
```

```python
x_train = vectorize_seq(train_data)
x_test = vectorize_seq(test_data)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
#model 

from keras import models 
from keras import layers 

model = models.Sequential() 
model.add(layers.Dense(64, activation='relu', input_shape=(10000,)))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(46, activation='softmax'))

model.compile(optimizer='rmsprop', loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
x_val = x_train[:1000]   #
p_x_train = x_train[1000:]   #

y_val = one_hot_train_labels[:1000]  # 
p_y_train = one_hot_train_labels[1000:]  # 
```

```python
#  

history = model.fit(p_x_train, p_y_train, epochs=20, 
                    batch_size=512, 
                    validation_data=(x_val, y_val))
```

```python
###  

history.history.keys()
```

```python
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(1, len(loss)+1)
```

```python
import matplotlib.pyplot as plt

plt.plot(epochs, loss, label='Train loss')
plt.plot(epochs, val_loss, label='Validation loss')
plt.legend()
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()
```

```python
np.argmin(val_loss)
```

```python
### epoch=8 

model = models.Sequential() 
model.add(layers.Dense(64, activation='relu', input_shape=(10000,)))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(46, activation='softmax'))

model.compile(optimizer='rmsprop', loss='categorical_crossentropy',
              metrics=['accuracy'])

model.fit(x_train, one_hot_train_labels, epochs=8,batch_size=512)
```

```python
model.evaluate(x_test, one_hot_test_labels)
```
